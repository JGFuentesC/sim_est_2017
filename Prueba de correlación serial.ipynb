```python
from __future__ import division
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
data = stats.uniform.rvs(size=1000)
```

```python
pd.Series(data).hist()
```

```python
def corr_serial(k,data):
    alpha = 0.05
    df = pd.DataFrame({'x':data})
    df['k'] = df.x.shift(-k)
    df.dropna(inplace=True)
    n = len(df)
    df['R']=1/(n-k)*(df.x-1/2)*(df.k-1/2)
    R=df.R.sum()
    return (R-stats.norm.ppf(1-alpha/2)/(12*np.sqrt(n-k)),R+stats.norm.ppf(1-alpha/2)/(12*np.sqrt(n-k)))
```

```python
for i in range(1,1000):
    print "k=%d | "%i,corr_serial(i,data)
```
